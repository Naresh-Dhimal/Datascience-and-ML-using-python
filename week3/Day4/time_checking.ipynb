```python
import timeit
```

```python
def timer(number, repeat):
    def wrapper(func):
        runs = timeit.repeat(func, number=number, repeat=repeat)
        print(sum(runs)/len(runs))
    return wrapper
```

```python
from timer import timer
@timer(1,1)
def main():
    for i in range(100):
      print(i)  
```
