## How does the average satisfaction level change with increasing years of employment?

```python
plt.figure(figsize=(6,4))
plt.title("Average satisfaction level change with increasing years of employment")
plt.plot(y_data, x_data,"o-", color="purple")
plt.xlabel("time_spend_company")
plt.ylabel("satisfaction_level")
plt.grid()
plt.show()
```

## How many employees have spent a specific number of years (time spent in the company)?

```python
count ={}
for i in df["time_spend_company"]:
    if i in count:
        count[i] += 1 
    else:
    
        count[i] = 1
count
```

```python
x = []
height = []
for key, value in count.items():
    
    x.append(key)
    height.append(value)
```

```python
# plotting box 
plt.title("Employees have spent a specific number of years")
plt.bar(x, height=height, color="purple")
plt.xlabel("time_spend_company")
plt.ylabel("no. of employees")
plt.show()
```

### What is the distribution of satisfaction levels among employees?

```python
# Calculate basic statistics
mean_satisfaction = df['satisfaction_level'].mean()
median_satisfaction = df['satisfaction_level'].median()
mode_satisfaction = df['satisfaction_level'].mode()[0]

print(f"Mean satisfaction level: {mean_satisfaction}")
print(f"Median satisfaction level: {median_satisfaction}")
print(f"Mode satisfaction level: {mode_satisfaction}")
```

```python
# Plot histogram to visualize the distribution
plt.title("Distribution of satisfaction levels among employees")
plt.hist(df['satisfaction_level'], bins=20, edgecolor='black')
plt.xlabel("satisfaction_level")
plt.ylabel("no. of employees")
plt.axvline(mean_satisfaction,color='red', label="Mean",linestyle='--')
plt.axvline(median_satisfaction, color='green', label="Medain", linestyle='--')
plt.axvline(mode_satisfaction, color='purple', label="Mode",linestyle='--')
plt.legend()
plt.show()
```

## From above, histogram shows Distribution of satisfaction levels among employees
* In x axis , satisfaction level is label
* In y axis, no. of employees is label
* Distribution of satisfaction levels among employees was left tail that means negatively skewed
* As histogram is negatively skewed, median is greater than mean

## Figure out employee leaving status (time_spend vs satisfaction_level vs left?

```python
time_spend = df.groupby(["time_spend_company"]).agg({"satisfaction_level": "mean", "left": "sum"}).reset_index()

x_data = time_spend["time_spend_company"]
y_data= time_spend["satisfaction_level"]
z_data = time_spend["left"]
z_weights = (z_data + z_data.min()) / (z_data.max() - z_data.min())



# Plot figure size 
plt.figure(figsize=(10, 6))

# Plot bubble plot
plt.scatter(x_data, y_data, s=z_weights * 300 + 20, c=z_weights, cmap='coolwarm')

# Set labels and title
plt.xlabel('Time Spent (years)')
plt.ylabel('Mean Satisfaction Level')
plt.title('Leaving department of employees by Time Spend in company and Satisfaction Level')
plt.colorbar(label='Number of Employees Left')

plt.grid()
plt.show()
```

### From above, bubble plot shows employee leaving status (time_spend vs satisfaction_level vs left)
* 2year time spend in company has high number of employee left even satisfaction level is approximately 0.63
* least number employees has left in the time spend of 6 years in a company

## How are the numerical variables correlated with each other?

```python
df.corr(numeric_only= True)
```

```python
plt.figure(figsize=(10, 7))

heatmap = plt.imshow(df.corr(numeric_only= True))

# Add color bar
plt.colorbar(heatmap, label='Correlation')

# Add labels and annotations
for i in range(len(df.corr(numeric_only= True).columns)):
    for j in range(len(df.corr(numeric_only= True).columns)):
        plt.text(j, i, f'{round(df.corr(numeric_only= True).iloc[i, j],3)}', ha='center', va='center', color='white')

# Set ticks and labels
plt.xticks(range(len(df.corr(numeric_only= True).columns)), df.corr(numeric_only=True).columns, rotation=45)
plt.yticks(range(len(df.corr(numeric_only= True).columns)), df.corr(numeric_only=True).columns)
plt.title("Numerical variables correlated with each other")
plt.show()
```

## From above , heatmap shows Numerical variables correlated with each other
* In above heatmap shows the correlation between numerical variables
* number poject and average montly hours has higher correlated among the department with value 0.417
* there is least correlated between left and satisfaction_level department with value -0.388

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Numerical variables correlated with each other")
plt.show()
```

## How to create different scatter plots under single matplotlib figure?

```python
fig, axis = plt.subplots(3,2, figsize=(10,16) )

# for [0,0]
# "satisfaction_level VS last_evaluation"
ax = axis[0,0]
x = df["satisfaction_level"]
y = df["last_evaluation"]
ax.scatter(x,y,color="red")
ax.set_title("satisfaction_level VS last_evaluation")
ax.set_xlabel("satisfaction_level")
ax.set_ylabel("last_evaluation")


# for [0,1]
# "satisfaction_level VS number_project"
ax = axis[0,1]
x = df["satisfaction_level"]
y = df["number_project"]
ax.scatter(x,y,color="green")
ax.set_title("satisfaction_level VS number_project")
ax.set_xlabel("satisfaction_level")
ax.set_ylabel("number_project")

# for[1,0]
# "satisfaction_level VS time_spend_company"
ax = axis[1,0]
x = df["satisfaction_level"]
y = df["time_spend_company"]
ax.scatter(x,y, color="yellow")
ax.set_title("satisfaction_level VS time_spend_company")
ax.set_xlabel("satisfaction_level")
ax.set_ylabel("time_spend_company")


# for [2,0]
# "last_evaluation VS number_project"
ax = axis[1,1]
x = df["last_evaluation"]
y = df["number_project"]
ax.scatter(x,y, color="purple")
ax.set_title("last_evaluation VS number_project")
ax.set_xlabel("satisfaction_level")
ax.set_ylabel("number_project")

# for [2,0]
# "last_evaluation VS time_spend_company"
ax = axis[2,0]
x = df["last_evaluation"]
y = df["time_spend_company"]
ax.scatter(x,y)
ax.set_title("last_evaluation VS number_project")
ax.set_xlabel("last_evaluation")
ax.set_ylabel("number_project")

# for [2,0]
# "last_evaluation VS time_spend_company"
ax = axis[2,1]
x = df["last_evaluation"]
y = df["average_montly_hours"]
ax.scatter(x,y)
ax.set_title("last_evaluation VS average_montly_hours")
ax.set_xlabel("last_evaluation")
ax.set_ylabel("average_montly_hours")

plt.show()

```

### From above, scatterplot shows different relation of columns in dataset
* satisfaction_level VS last_evaluation & last_evaluation VS average_montly_hours has scattered values
* where other columns relation has approximately linearly distribute
