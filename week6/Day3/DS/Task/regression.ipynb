##  Simple Linear Regression (y= 2x + 3)

```python
# let's plot graphs for y=2x + 3 equations using independent variables ranging from 100 to 200.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
# generating x value using numpy random with size 10
x_values = np.random.randint(low=100,high=200,size=10)
```

```python
# displaying x values
x_values
```

```python
# calculating y values from given equations
y_values = 2 * x + 3
```

```python
# displaying y values
y_values
```

```python
# Define the equation text
equation_text ="2x + 3"

#resizing the figure size
plt.figure(figsize=(8,6))

# plotting Line
plt.plot(x_values, y_values,"o-", color="purple", label=equation_text)

# Annotate each data point with its coordinates
for i in range(len(x_values)):
    plt.text(x_values[i],y_values[i],f"({x_values[i]},{y_values[i]})")

# Add title and labels
plt.title("Line plot of equation")
plt.text(160,315, equation_text, fontsize=15, color="red") 
plt.xlabel("X_values")
plt.ylabel("y_values")

# Enable grid and legend
plt.grid()
plt.legend()

# Display the plot
plt.show()
```

## Multiple Linear Regression 

```python

# generating x1,x2,x3
np.random.seed(0)
x1 = np.random.randint(low=100,high=200,size=5)
x2 = np.random.randint(low=100,high=200,size=5)
x3 = np.random.randint(low=100,high=200,size=5)
```

```python
# displaying x1,x2,x3 values
print(x1)
print(x2)
print(x3)
```

```python
# calculating y value
y = 3 * x1 + 2 * x2 + 4 * x3 + 5
```

```python
# displaying y values
print(y)
```

```python
# plotting the line
plt.plot(x1,y,"o-")
plt.plot(x2,y)
plt.plot(x3,y)
```

## Polynomial Regression

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(1)
# generating x value using numpy random with size 10
x = np.random.randint(low=100,high=200,size=5)
```

```python
# Displaying x values
print(x)
```

```python
# calculating y value
y = 2 * x**2 + 3 * x + 4
```

```python
# displaying y values
print(y)
```

```python
#resizing the figure size
plt.figure(figsize=(8,6))

# plotting Line
plt.plot(x, y,"o-", color="purple")

# Annotate each data point with its coordinates
for i in range(len(x)):
    plt.text(x[i],y[i],f"({x[i]},{y[i]})")

# Add title and labels
plt.title("Line plot of equation")
plt.xlabel("X_values")
plt.ylabel("y_values")

# Enable grid 
plt.grid()

# Display the plot
plt.show()
```
