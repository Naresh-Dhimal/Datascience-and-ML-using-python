# If we have another numpy array of shape (8, 4341). Can we perform any operations on them? Is it broadcastable?

-> yes we can perform operation  of another array with shape(8, 4341) and it is broadcastable because to perform matrix multiplication, the first matrix must have the same number of columns as the second matrix has rows.

```python
# for example
a = np.array([[12,12,12],[12,34,4343]])
b = np.array([[12,34,12,],[12,321,14],[3,1,4]])
```

```python
a.shape
```

```python
b.shape
```

# here, array a has same number of columns as number of rows of b so, a and b can perform an operation and broadcastable

```python
c = np.dot(a,b)
print(c)
```

# Applying Universal Functions and Aggregations

```python
#removing zero indes data
data_set = [dataset[index] for index in range(1,len(dataset))]
selling_price =[item[2] for item in data_set]
selling_price_array = np.array(selling_price, dtype =float)
print(type(selling_price))
print(selling_price_array)
```

# applying universal function
<b>.sin(), .cos(), .tan()

```python

# Sine
sin_value = np.sin(selling_price_array)
print(f"Sine : {sin_value}")

# Cosine
cosin_value = np.cos(selling_price_array)
print(f"Cosine : {cosin_value}")

# tangent
tan_value = np.tan(selling_price_array)
print(f"Tangent : {tan_value}")
```

```python
# Square root
sqrt_ans = np.sqrt(selling_price_array)
print('Square Root:', sqrt_ans)
```

# Aggregation

```python
# mean()
np.mean(selling_price_array)
```

```python
# median()
np.median(selling_price_array)
```

```python
# min()
np.min(selling_price_array)
```

```python
# argmin()
np.argmin(selling_price_array)
```

```python
# max()
np.max(selling_price_array)
```

```python
# argmax()
np.argmax(selling_price_array)
```
