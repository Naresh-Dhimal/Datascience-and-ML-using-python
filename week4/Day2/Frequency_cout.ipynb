```python
# sample data
colors = [
    "Red", "Blue", "Green", "Red", "Yellow", "Blue", "Green", "Green", "Blue", "Red",
    "Red", "Blue", "Yellow", "Green", "Green", "Blue", "Red", "Red", "Red", "Blue",
    "Green", "Yellow", "Blue", "Blue", "Green", "Red", "Red", "Green", "Blue", "Green",
    "Yellow", "Red", "Blue", "Green", "Blue", "Red", "Blue", "Green", "Blue", "Red",
    "Green", "Green", "Yellow", "Red", "Blue", "Green", "Blue", "Red", "Green", "Red"
]
```

```python
colors_counts= Counter(colors)
print(colors_counts)
```

```python
print("Frequency Table")
print("Color Frequency")
print("------------------")

for color, count in colors_counts.items():
  print(f"{color}  {count}")
```

```python
colors_counts = dict(colors_counts)
index_list = list(colors_counts.keys())
value_list = list(colors_counts.values())
```

```python
df_color = pd.DataFrame(value_list, index_list , columns= ["Frequency"])
```

```python
df_color
```

```python
df_color.plot(kind='bar')
```

```python
# Sample data - list of ages
ages = [
    28, 34, 25, 28, 22, 31, 25, 27, 30, 24,
    22, 29, 26, 31, 28, 34, 30, 25, 27, 26,
    23, 29, 32, 28, 26, 24, 30, 25, 28, 31,
    33, 27
]
print(ages)
```

```python
ages.sort()
ages
```

```python
interval = [(20, 24),(25, 30),(31, 34),(35,40)]
```

```python
interval_counts = np.zeros(len(interval), dtype=int)
for age in ages:
  for i, (lower, upper) in enumerate(interval):
    if lower <= age <= upper:
      interval_counts[i] += 1
```

```python
interval_counts
```

```python
print("Frequency Table for Age Intervals:")
print("Interval  Frequency")
print("____________________")
for (lower,upper), count in zip(interval, interval_counts):
  print(f"{lower}-{upper}       {count} ")
```

```python
interval_labes = [f"{lower}-{upper}" for (lower, upper) in interval]
```

```python
plt.bar(x=interval_labes, height = interval_counts, color='skyblue')
```
