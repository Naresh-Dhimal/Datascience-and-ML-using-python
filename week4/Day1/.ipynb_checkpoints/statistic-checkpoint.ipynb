```python
import numpy as np
from scipy.stats import mode
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
```

# Sample Datat 1

```python
data1 = np.random.normal(loc=5,scale=4, size=50)
```

```python
data1.sort()
```

```python
data1_mean = np.mean(data1) # mean of data1
```

```python
data1_median = np.median(data1) # median of data1
```

```python
data1_mode = mode(data1)
```

```python
skew(data1)
# here, data1 is negative skewed
```

```python
sns.kdeplot(data1)
plt.axvline(5, color='red')
plt.axvline(data1_median, color='blue', linestyle="--")
plt.axvline(data1_mode.mode, color='black')
plt.show()
```

# Sample data2

```python
data2 = np.random.uniform(low=1, high= 50, size= 100)
```

```python
data2.sort()
```

```python
data2_median = np.median(data2)
```

```python
data2_mode = mode(data2)
```

```python
skew(data2)
```

```python
sns.kdeplot(data2)
plt.axvline(5,color="red")
plt.axvline(data2_median, color= "blue", linestyle= "--")
plt.axvline(data2_mode.mode, color= "black")
plt.show()
```

# Sample data 3

```python
data3 = list(np.random.randint(1,20,size= 50))
data3.append(100)
```

```python
data3.sort()
```

```python
skew(data3)
```

```python
# positive skewed
```

```python
data3_mean = np.mean(data3)
```

```python
data3_median = np.median(data3)
```

```python
data3_mode = mode(data3)
```

```python
sns.kdeplot(data3)
plt.axvline(data3_mean,color= "blue")
plt.axvline(data3_median,color= "red")
plt.axvline(data3_mode.mode,color= "black")
plt.show()
```

```python
# here, mean > median > mode hence, given data3 is right skewed distribution
```
