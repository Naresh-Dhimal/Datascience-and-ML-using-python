```python
def operation(a,b):
    oper = input("Enter a operation('+','-','*','/','%','**','//'):")
    if oper == '+':
        print(f"Addition operation of {a} and {b}")
        total_sum = a + b
        return total_sum
    if oper == '-':
        print(f"Substraction operation of {a} and {b}")
        diff = a - b
        return diff
    if oper == '*':
        print(f"Multiplication operation of {a} and {b}")
        mul = a * b
        return mul
    if oper == '/':
        print(f"Division operation of {a} and {b}")
        div = a / b
        return div
    if oper == '%':
        print(f"Modules operation of {a} and {b}")
        mod = a % b
        return mod
    if oper == '**':
        print(f"Exponential operation of {a} and {b}")
        expo = a ** b
        return expo
    if oper == '//':
        print(f"Floor Division of {a} and {b}")
        floor = a // b
        return floor
    else:
        print("Invalid operation.")
        
```

```python
num1 = int(input("Enter a first number:"))
num2 = int(input("Enter a second number:"))
result = operation(a=num1, b=num2)
print(result)
```
